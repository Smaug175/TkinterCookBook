```python
from tkinter import *

def printSelection():
    num=var.get()
    if num==1:
        label.config(text='boy')
    else:
        label.config(text='girl')
    
root=Tk()
root.title('ch7_1')

var=IntVar()
var.set(1)

label=Label(root,text='None',bg='yellow',width=30)
label.pack()

rbman=Radiobutton(root,text='boy',
                  variable=var,value=1,
                  command=printSelection)
rbman.pack(anchor=W)
rbwoman=Radiobutton(root,text='girl',
                    variable=var,value=2,
                    command=printSelection)
rbwoman.pack(anchor=W)

root.mainloop()
```

```python
#但是当选项变多时程序就会显得比较复杂，
# 此时可以考虑使用字典存储选项按钮相关信息，
# 然后用遍历字典方式建立选项按钮，
# 可参考下列实例
from tkinter import *

def printSelection():
    print(cities[var.get()])

root=Tk()
root.title('ch7_3')
cities={0:'New York',1:'Paris',2:'London',3:'Beijing',4:'Tokyo'}

var=IntVar()
var.set(0)
label=Label(root,text='select the most favorite city ',bg='lightyellow',fg='blue',
            width=30).pack()

for val,city in cities.items():
    Radiobutton(root,
                text=city,
                variable=var,value=val,
                command=printSelection).pack(anchor=W)
    #Radiobutton表示单选按钮
root.mainloop()
```

```python
from tkinter import *

def printSelection():
    print(cities[var.get()])

root=Tk()
root.title('ch7_3')
cities={0:'New York',1:'Paris',2:'London',3:'Beijing',4:'Tokyo'}

var=IntVar()
var.set(0)
label=Label(root,text='select the most favorite city ',bg='lightyellow',fg='blue',
            width=30).pack()

for val,city in cities.items():
    Radiobutton(root,
                text=city,
                indicatoron=0,#盒子取代按钮选项
                width=30,
                variable=var,value=val,
                command=printSelection).pack()
    #Radiobutton表示单选按钮
root.mainloop()
```

```python
from tkinter import *

def printInfo():
    selection=''
    for i in checkboxes:
        if checkboxes[i].get()==True:
            selection= selection+sports[i]+'\t'
    print(selection)

root=Tk()
root.title('ch7_8')

lab=Label(root,text='Which do you like best?',
          fg='blue',bg='lightyellow',width=30)
lab.grid(row=0,column=0)

sports={0:'NFL',1:'MLB',2:'NBA'}
checkboxes={}#用来储存复选框对象
for i in range(3):
    checkboxes[i]=BooleanVar()
    Checkbutton(root,text=sports[i],variable=checkboxes[i]).grid(row=i+1,column=0,sticky=W)

btn=Button(root,text='Info',width=10,command=printInfo)
btn.grid(row=i+2,column=0)

root.mainloop()
```

```python
from tkinter import *   

def selAll():
    entry.select_range(0,END)

def deSel():
    entry.select_clear()

def clr():
    entry.delete(0,END)

def readonly():
    if var.get()==True:
        entry.config(state=DISABLED)
    else:
        entry.config(state=NORMAL)
        
root=Tk()
root.title('ch7_9')

entry=Entry(root,width=30)
entry.grid(row=0,column=0,columnspan=4,
           padx=5,pady=5,sticky=W)

btnSel=Button(root,text='Select All',width=8,command=selAll)
btnSel.grid(row=1,column=0,sticky=W,padx=5,pady=5)
btnDesel=Button(root,text='Deselect',width=8,command=deSel)
btnDesel.grid(row=1,column=1,sticky=W,padx=5,pady=5)
btnClr=Button(root,text='Clear',width=8,command=clr)
btnClr.grid(row=1,column=2,sticky=W,padx=5,pady=5)
btnQuit=Button(root,text='Quit',width=8,command=root.destroy)
btnQuit.grid(row=1,column=3,sticky=W,padx=5,pady=5)

var=BooleanVar()
var.set(False)
chkReadonly=Checkbutton(root,text='Readonly',variable=var,command=readonly)
chkReadonly.grid(row=2,column=0,sticky=W,padx=5,pady=5)

root.mainloop()
```
