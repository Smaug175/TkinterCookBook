快捷键是某个菜单类别或是列表指令的英文字符串内为单一字母增加下画线，然后可以用Alt键先启动此功能，当菜单显示下画线字母时，可以直接按指定字母键启动该功能。设计方式是在下列两个方法内增加underline参数。

```python
from tkinter import *
from tkinter import messagebox

def newfile():
    messagebox.showinfo("New File", "New File")
def openfile():
    messagebox.showinfo("Open File", "Open File")
def savefile():
    messagebox.showinfo("Save File", "Save File")
def saveasfile():
    messagebox.showinfo("Save As File", "Save As File") 

def about():
    messagebox.showinfo("About", "This is a simple example of a menu")
    
root=Tk()
root.title("Hello World")
root.geometry("300x200")

menubar=Menu(root)
filemenu=Menu(menubar, tearoff=0)
menubar.add_cascade(label="File", menu=filemenu,underline=0)#添加子菜单

filemenu.add_command(label="New", command=newfile,underline=0)
filemenu.add_command(label="Open", command=openfile,underline=0)
filemenu.add_separator()#添加分割线
filemenu.add_command(label="Save", command=savefile,underline=0)
filemenu.add_command(label="Save As", command=saveasfile,underline=5)
filemenu.add_separator()#添加分割线
filemenu.add_command(label='exit', command=root.destroy,underline=1)

helpmenu=Menu(menubar, tearoff=0)
menubar.add_cascade(label="Help", menu=helpmenu,underline=0)#添加子菜单
helpmenu.add_command(label="About", command=about,underline=1)

root.config(menu=menubar)#显示菜单

root.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox

def newfile():
    messagebox.showinfo("New File", "New File")
def openfile():
    messagebox.showinfo("Open File", "Open File")
def savefile():
    messagebox.showinfo("Save File", "Save File")
def saveasfile():
    messagebox.showinfo("Save As File", "Save As File") 

def findNext():
    messagebox.showinfo("Find Next", "Find Next")
def findPre():
    messagebox.showinfo("Find Pre", "Find Pre")

def about():
    messagebox.showinfo("About", "This is a simple example of a menu")
    
root=Tk()
root.title("Hello World")
root.geometry("300x200")

menubar=Menu(root)
filemenu=Menu(menubar, tearoff=0)
menubar.add_cascade(label="File", menu=filemenu,underline=0)#添加子菜单

filemenu.add_command(label="New", command=newfile,underline=0,accelerator="Ctrl+N")
filemenu.add_command(label="Open", command=openfile,underline=0)
filemenu.add_separator()#添加分割线
filemenu.add_command(label="Save", command=savefile,underline=0)
filemenu.add_command(label="Save As", command=saveasfile,underline=5)
filemenu.add_separator()#添加分割线
filemenu.add_command(label='exit', command=root.destroy,underline=1)
filemenu.add_separator()#添加分割线
findmenu=Menu(filemenu, tearoff=0)
filemenu.add_cascade(label="Find", menu=findmenu,underline=0)#添加子菜单
findmenu.add_command(label="Find Next", command=findNext,underline=0)
findmenu.add_command(label="Find Pre", command=findPre,underline=0)

helpmenu=Menu(menubar, tearoff=0)
menubar.add_cascade(label="Help", menu=helpmenu,underline=0)#添加子菜单
helpmenu.add_command(label="About", command=about,underline=1)

root.config(menu=menubar)#显示菜单
root.bind("<Control-n>", lambda event:newfile())#绑定快捷键

root.mainloop()
```

```python
from tkinter import *
from tkinter import messagebox

def minimizeicon():
    root.iconify()
def showpopupmenu(event):
    popupmenu.post(event.x_root, event.y_root)

root=Tk()
root.title("Hello World")
root.geometry("300x200")

popupmenu=Menu(root, tearoff=0)
popupmenu.add_command(label="Minimize", command=minimizeicon)
popupmenu.add_separator()#添加分割线   
popupmenu.add_command(label="Exit", command=root.destroy)
root.bind("<Button-3>", showpopupmenu)#绑定鼠标右键

root.mainloop()
```

第7行的post( )方法是由popupmenu对象启动，相当于可以在鼠标光标位置（event.x_root,event.y_root）弹出此菜单。

```python
from tkinter import *

def status():
    if demostatus.get():
        statusbar.pack(side=BOTTOM, fill=X)#显示状态栏
    else:
        statusbar.pack_forget()#隐藏状态栏

root=Tk()
root.title("Hello World")
root.geometry("300x200")

menubar=Menu(root)

filemenu=Menu(menubar, tearoff=0)
menubar.add_cascade(label="File", menu=filemenu)#添加子菜单 
filemenu.add_command(label="exit", command=root.destroy)

viewmenu=Menu(menubar, tearoff=0)
menubar.add_cascade(label="View", menu=viewmenu)#添加子菜单
demostatus=BooleanVar()
demostatus.set(True)
viewmenu.add_checkbutton(label="Status", 
                         command=status, variable=demostatus)#添加复选框
root.config(menu=menubar)

statusVar=StringVar()
statusVar.set("This is a status bar")
statusbar=Label(root, textvariable=statusVar,
                bd=1, relief=SUNKEN, anchor=W)
statusbar.pack(side=BOTTOM, fill=X)#显示状态栏

root.mainloop()
```

这个程序会建立一个File菜单，菜单内有Exit命令。这个程序也建立了一个工具栏，在工具栏内有exitBtn按钮。这个程序不论是执行File菜单的Exit命令或是单击工具栏中的exitBtn按钮，都可以让程序结束。

```python
from tkinter import *

root=Tk()
root.title("Hello World")
root.geometry("300x200")

menubar=Menu(root)
filemenu=Menu(menubar, tearoff=0)
menubar.add_cascade(label="File", menu=filemenu)#添加子菜单
filemenu.add_command(label="exit", command=root.destroy)

toolbar=Frame(root, height=20, bg="lightgray",relief=SUNKEN,borderwidth=3)

sungif=PhotoImage(file="D:\Desk\GUI\TkinterCookBook\img\\1.gif")
exitbutton=Button(toolbar, image=sungif, command=root.destroy)
exitbutton.pack(side=LEFT, padx=2, pady=2)
toolbar.pack(side=TOP, fill=X)
root.config(menu=menubar)

root.mainloop()
```
