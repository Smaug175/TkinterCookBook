```python
from tkinter import *
import math 

tk=Tk()
canvas=Canvas(tk,width=800,height=600)
canvas.pack()

x_center,y_center,r=320,240,100

x,y=[],[]
for i in range(24):
    x.append(x_center+r*math.cos(2*math.pi*i/12))
    y.append(y_center+r*math.sin(2*math.pi*i/12))
for i in range(24):
    for j in range(24):
        canvas.create_line(x[i],y[i],x[j],y[j])
        
tk.mainloop()
```

```python
from tkinter import *
import time
def bllmove(event):
    if event.keysym=='Up':
        canvas.move(1,0,-3)
    elif event.keysym=='Down':
        canvas.move(1,0,3)
    elif event.keysym=='Left':
        canvas.move(1,-3,0)
    elif event.keysym=='Right':
        canvas.move(1,3,0)
        
tk=Tk()
canvas=Canvas(tk,width=800,height=600) 
canvas.pack()
canvas.create_polygon(10,10,10,60,50,35)
canvas.bind_all('<KeyPress-Up>',bllmove)
canvas.bind_all('<KeyPress-Down>',bllmove)
canvas.bind_all('<KeyPress-Left>',bllmove)
canvas.bind_all('<KeyPress-Right>',bllmove)
mainloop()
```
