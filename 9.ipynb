```python
from tkinter import *

def bgupgrate(source):
    red=r.get()
    green=g.get()
    blue=b.get()
    print("red=%d,green=%d,blue=%d" % (red,green,blue))
    mycolor="#%02x%02x%02x" % (red,green,blue)
    root.config(bg=mycolor)

root=Tk()
root.title("ch9_4")
root.geometry("300x160")

r=Scale(root,from_=0,to=255,label="red",command=bgupgrate)
g=Scale(root,from_=0,to=255,label="green",command=bgupgrate)
b=Scale(root,from_=0,to=255,label="blue",command=bgupgrate)
g.set(125)
r.grid(row=0,column=0)
g.grid(row=0,column=1)
b.grid(row=0,column=2)

root.mainloop()
```

上述设计是将尺度条放置在窗口左上角，如果想调整位置并不太方便，最好的设计方式是先设计一个容器，然后将这三个尺度条放置在此容器内，未来如果想要移动位置，可以直接移动容器位置。

```python
from tkinter import *
from tkinter.colorchooser import *

def bgupgrate():
    mycolor=askcolor()
    print(type(mycolor),mycolor)
    root.config(bg=mycolor[1])
    

root=Tk()
root.title("ch9_4_1")
root.geometry("300x160")

btn=Button(root,text='选择颜色',command=bgupgrate)
btn.pack(pady=5)

root.mainloop()
```

```python
from tkinter import *

def bgupgrate(source):
    red=r.get()
    green=g.get()
    blue=b.get()
    print("red=%d,green=%d,blue=%d" % (red,green,blue))
    mycolor="#%02x%02x%02x" % (red,green,blue)
    root.config(bg=mycolor)

root=Tk()
root.title("ch9_4")
root.geometry("300x160")

fm=Frame(root)
fm.pack()

r=Scale(fm,from_=0,to=255,label="red",command=bgupgrate)
g=Scale(fm,from_=0,to=255,label="green",command=bgupgrate)
b=Scale(fm,from_=0,to=255,label="blue",command=bgupgrate)
g.set(125)
r.grid(row=0,column=0)
g.grid(row=0,column=1)
b.grid(row=0,column=2)

root.mainloop()
```

```python
from tkinter import *

def printinf():
    print(sp.get())

root = Tk()
root.title("ch9_7")

sp=Spinbox(root,from_=0,to=100,increment=5,command=printinf)
sp.pack(pady=20,padx=10)

root.mainloop()
```

```python
#其实在使用Spinbox时也可以不设置初值和终值，而是将数值存储在序列数据中，然后使用values参数来指定序列数据。

from tkinter import *

def printinf():
    print(sp.get())
    
root = Tk()
root.title("ch9_8")

sp=Spinbox(root,
           values=(1,2,4,8,16,32,64,128,256,512),
           command=printinf)
sp.pack(pady=20,padx=10)

root.mainloop()
```
