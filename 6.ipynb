```python
'''
# 6.1
显示一个标签和一个按钮，按钮按下后，标签显示一条消息，再按下后消息消失
'''
from tkinter import *

def btn_bit():
    print("Button clicked")
    
    global msg_on
    if msg_on==False:
        msg_on=True
        x.set('I like tkinter')
    else:
        msg_on=False
        x.set('')

root=Tk()
root.title('ch6_1')

msg_on=False
x=StringVar()

label=Label(root,textvariable=x,
            fg='blue',bg='lightyellow',
            font='Times 16 bold',width=25,height=2)

label.pack()

btn=Button(root,text='Click Me',command=btn_bit)
btn.pack()

root.mainloop()
```

```python
'''
# 6.1
显示一个标签和一个按钮，按钮按下后，标签显示一条消息，再按下后消息消失
'''
from tkinter import *

def btn_bit():
    if x.get()=='':
        x.set('I like tkinter')
    else:
        x.set('')

root=Tk()
root.title('ch6_2')

x=StringVar()

label=Label(root,textvariable=x,
            fg='blue',bg='lightyellow',
            font='Times 16 bold',width=25,height=2)

label.pack()

btn=Button(root,text='Click Me',command=btn_bit)
btn.pack()

root.mainloop()
```

```python
from tkinter import *

def callback(*args):
    print("variable changed:",xE.get())

root=Tk()
root.title('ch6_3')

xE=StringVar()
entry=Entry(root,textvariable=xE)
entry.pack(padx=10,pady=10)
xE.trace("w",callback)#w表示变量xE被写入时，调用callback函数

root.mainloop()
```

```python
from tkinter import *

def callbackW(*args):
    xL.set(xE.get())

def callbackR(*args):
    print("waring! data is been reading")

def hit():
    print("read data:",xE.get())

root=Tk()
root.title('ch6_5')

xE=StringVar()

entry=Entry(root,textvariable=xE)
entry.pack(padx=10,pady=10)
xE.trace("w",callback)#w表示变量xE被写入时，调用callback函数,写入追踪
xE.trace("r",callbackR)#r表示变量xE被读取时，调用callbackR函数，读取追踪

xL=StringVar()
label=Label(root,textvariable=xL)
xL.set('synchronous display')
label.pack(padx=10,pady=10)

btn=Button(root,text='Read Data',command=hit)
btn.pack(padx=10,pady=10)

root.mainloop()
```

```python
#计算器的完整设计
from tkinter import *

def calculate():
    result=eval(equ.get())
    equ.set(equ.get()+'=\n'+str(result))

def show(buttonString):
    content=equ.get()
    if content=='0':
        content=''
    equ.set(content+buttonString)

def backspace():
    equ.set(equ.get()[:-1])

def clear():
    equ.set('0')

root=Tk()
root.title('calculator')

equ=StringVar()
equ.set('0')

lab=Label(root,text='',width=25,height=2,relief='raised',
          anchor=SE,textvariable=equ)
lab.grid(row=0,column=0,columnspan=4,padx=5,pady=5)

clearButton=Button(root,text='C',width=5,command=clear,fg='blue')
clearButton.grid(row=1,column=0,padx=5,pady=5)

Button(root,text='DEL',width=5,command=backspace,fg='blue').grid(row=1,column=1,padx=5,pady=5)
Button(root,text='/',width=5,command=lambda:show('/'),fg='blue').grid(row=1,column=2,padx=5,pady=5)
Button(root,text='*',width=5,command=lambda:show('*'),fg='blue').grid(row=1,column=3,padx=5,pady=5)

Button(root,text='7',width=5,command=lambda:show('7')).grid(row=2,column=0,padx=5,pady=5)
Button(root,text='8',width=5,command=lambda:show('8')).grid(row=2,column=1,padx=5,pady=5)
Button(root,text='9',width=5,command=lambda:show('9')).grid(row=2,column=2,padx=5,pady=5)
Button(root,text='-',width=5,command=lambda:show('-'),fg='blue').grid(row=2,column=3,padx=5,pady=5)

Button(root,text='4',width=5,command=lambda:show('4')).grid(row=3,column=0,padx=5,pady=5)
Button(root,text='5',width=5,command=lambda:show('5')).grid(row=3,column=1,padx=5,pady=5)
Button(root,text='6',width=5,command=lambda:show('6')).grid(row=3,column=2,padx=5,pady=5)
Button(root,text='+',width=5,command=lambda:show('+'),fg='blue').grid(row=3,column=3,padx=5,pady=5)

Button(root,text='1',width=5,command=lambda:show('1')).grid(row=4,column=0,padx=5,pady=5)
Button(root,text='2',width=5,command=lambda:show('2')).grid(row=4,column=1,padx=5,pady=5)
Button(root,text='3',width=5,command=lambda:show('3')).grid(row=4,column=2,padx=5,pady=5)

Button(root,text='0',width=5,command=lambda:show('0')).grid(row=5,column=0,padx=5,pady=5)
Button(root,text='.',width=5,command=lambda:show('.')).grid(row=5,column=1,padx=5,pady=5)
Button(root,text='=',width=5,command=calculate,fg='red',bg='yellow').grid(row=5,column=2,padx=5,pady=5)
Button(root,text='(',width=5,command=lambda:show('('),fg='blue').grid(row=4,column=3,padx=5,pady=5)
Button(root,text=')',width=5,command=lambda:show(')'),fg='blue').grid(row=5,column=3,padx=5,pady=5)

root.mainloop()







root.mainloop()
```
